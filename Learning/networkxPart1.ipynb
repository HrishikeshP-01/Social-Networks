```python
import networkx as nx
```

```python
G=nx.Graph()
```

```python
G.add_node(1)
G.add_node(3)
G.add_node(6)
G.add_node(8)
```

```python
G.add_edge(1,3)
G.add_edge(1,6)
G.add_edge(3,8)
G.add_edge(3,6)
G.add_edge(6,8)
```

```python
G.edges()
```

```python
import matplotlib.pyplot as plt
nx.draw(G)
plt.show()
```

```python
import matplotlib.pyplot as plt
nx.draw(G, with_labels=1)
plt.show()
```
